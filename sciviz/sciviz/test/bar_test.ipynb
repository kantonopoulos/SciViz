```python
import matplotlib.pyplot as plt
import seaborn as sns
import sciviz
```

```python
iris = sns.load_dataset('iris')
iris['size'] = iris['sepal_length'].apply(lambda x: 'big' if x > 5.5 else 'small')
```

Simple barplot with x and y variables.

```python
ax = sciviz.bar(iris, 'species', 'sepal_length')
```

Add color variable.

```python
ax = sciviz.bar(iris, 'species', 'sepal_length', color='species', alpha = 1)
```

Add another color variable and color palette.

```python
ax = sciviz.bar(iris, 'species', 'sepal_length', color='size', color_pal=['#FEED70', '#603479'])
```

Change stat and do count plot.

```python
ax = sciviz.bar(iris, 'species', 'sepal_length', stat='count')
```

Change color order.

```python
ax = sciviz.bar(iris, 'species', 'sepal_length', color='species', color_order=['virginica', 'versicolor', 'setosa'])
```

Change order and color order.

```python
ax = sciviz.bar(iris, 'species', 'sepal_length', color='species', order=['virginica', 'versicolor', 'setosa'], 
         color_order=['virginica', 'versicolor', 'setosa'])
```

Change fill and width.

```python
ax = sciviz.bar(iris, 'species', 'sepal_length', color='species', width=0.8, fill=False, edgecolor=None, legend=None)
```

Add red edgecolor and errorbars.

```python
ax = sciviz.bar(iris, 'species', 'sepal_length', fill=False, edgecolor='red', 
         errorbar = sciviz.error_parameters())
```

Customize errorbars.

```python
ax = sciviz.bar(iris, 'species', 'sepal_length', fill=False, edgecolor='red', 
         errorbar = sciviz.error_parameters(errorbar=('sd', 2), color_pal=['red'], linestyle='--', linewidth=2, capsize=0.1))
```

Change orientation and alpha.

```python
ax = sciviz.bar(iris, 'sepal_length', 'species', color='species', orient='h', alpha=0.3, legend=None)
```
