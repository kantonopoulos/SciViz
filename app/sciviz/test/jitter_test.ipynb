```python
import matplotlib.pyplot as plt
import seaborn as sns
from fun import jitter, crossbar_parameters
```

```python
iris = sns.load_dataset('iris')
iris['size'] = iris['sepal_length'].apply(lambda x: 'big' if x > 5.5 else 'small')
```

```python
ax = jitter(iris, 'sepal_length', 'species', color='species', orient='h', crossbar=crossbar_parameters(barstyle='|'))
```

Simple jitter with x and y variables.

```python
ax = jitter(iris, 'species', 'sepal_length')
```

Add color variable.

```python
ax = jitter(iris, 'species', 'sepal_length', color='species')
```

Add color based on a different variable with and without dodge and different color palette.

```python
ax = jitter(iris, 'species', 'petal_length', color='size', color_pal=['#FFFF80', '#603479'])
```

```python
ax = jitter(iris, 'species', 'petal_length', color='size', dodge=True, color_pal=['#FFFF80', '#603479'])
```

Change color order.

```python
ax = jitter(iris, 'species', 'sepal_length', color='species', color_order=['virginica', 'versicolor', 'setosa'])
```

Change color order and order.

```python
ax = jitter(iris, 'species', 'sepal_length', color='species', order=['virginica', 'versicolor', 'setosa'],
            color_order=['virginica', 'versicolor', 'setosa'])
```

Change marker size and alpha. Remove legend.

```python
ax = jitter(iris, 'species', 'sepal_length', color='species', size=100, alpha=0.5, legend=None)
```

Change orientation and jitter (False or number).

```python
ax = jitter(iris, 'sepal_length', 'species', color='species', orient='h', jitter=False)
```

```python
ax = jitter(iris, 'sepal_length', 'species', color='species', orient='h', jitter=0.3)
```

Add crossbars.

```python
ax = jitter(iris, 'species', 'sepal_length', color='species', 
            crossbar=crossbar_parameters())
```

Add crossbars that are dodged.

```python
ax = jitter(iris, 'species', 'petal_length', color='size', color_pal=['#FFFF80', '#603479'], dodge=True, 
            crossbar=crossbar_parameters(color_val='size', color_pal=['black'], barstyle='_', barsize=20, barwidth=3))
```

Customize crossbars.

```python
ax = jitter(iris, 'species', 'sepal_length', color_pal=['black'], size=100, alpha=0.5,
            crossbar=crossbar_parameters(color_val='size', color_pal=['blue', 'red'], barstyle='o', barsize=20, barwidth=3))
```
