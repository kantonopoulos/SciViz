```python
import matplotlib.pyplot as plt
import seaborn as sns
from fun import venn, label_parameters
```

```python
iris = sns.load_dataset('iris')
iris['size'] = iris['sepal_length'].apply(lambda x: 'big' if x > 5.5 else 'small')
```

Simple Venn plot with 3 sets.

```python
venn(iris, 'sepal_length', 'species', color_pal=None)
```

Simple Venn plot with 2 sets. Change palette.

```python
venn(iris, 'petal_length', 'size', color_pal=['#FFFF80', '#603479'])
```

Customize labels.

```python
venn(iris, 'sepal_length', 'species', color_pal=None, 
labels=label_parameters(size=14, color='grey'))
```

Remove labels.

```python
venn(iris, 'sepal_length', 'species', color_pal=None, labels=None)
```
