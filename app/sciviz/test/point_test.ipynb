```python
import matplotlib.pyplot as plt
import seaborn as sns
from fun import point, legend_parameters
```

```python
iris = sns.load_dataset('iris')
iris['size'] = iris['sepal_length'].apply(lambda x: 'big' if x > 5.5 else 'small')
```

Simple scatterplot with x and y variables.

```python
ax = point(iris, 'sepal_width', 'sepal_length')
```

Change color, shape and size of markers.

```python
ax = point(iris, 'sepal_width', 'sepal_length', color_pal='red', shape_pal='x', size=80)
```

Add color variable.

```python
ax = point(iris, 'sepal_width', 'sepal_length', color='species')
```

Add another variable for shape and size.

```python
ax = point(iris, 'sepal_width', 'sepal_length', color='species', shape='size', size='size', 
           legend=legend_parameters(title=['Species', 'Size']))
```

Change color and shape order. Use the same variable in both.

```python
ax = point(iris, 'sepal_width', 'sepal_length', color='species', shape='species', color_order=['virginica', 'versicolor', 'setosa'])
```

Remove legend, change shape, size and alpha.

```python
ax = point(iris, 'sepal_width', 'sepal_length', color='species', alpha=1, size=30, shape_pal='^', legend=None)
```
