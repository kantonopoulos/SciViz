```python
import matplotlib.pyplot as plt
import seaborn as sns
from fun import violin, box_parameters
```

```python
iris = sns.load_dataset('iris')
iris['size'] = iris['sepal_length'].apply(lambda x: 'big' if x > 5.5 else 'small')
```

Simple violinplot with x and y variables.

```python
ax = violin(iris, 'species', 'sepal_length')
```

Add color variable.

```python
ax = violin(iris, 'species', 'sepal_length', color='species')
```

Add another color variable and increase width.

```python
ax = violin(iris, 'species', 'petal_width', color='size', width=0.8)
```

Enable split, change palette and remove legend.

```python
ax = violin(iris, 'species', 'petal_length', color='size', color_pal=['#FFFF80', '#603479'], split=True, width=0.8, legend=None)
```

Add color order.

```python
ax = violin(iris, 'species', 'sepal_length', color='species', color_order=['virginica', 'versicolor', 'setosa'])
```

Add color order and order.

```python
ax = violin(iris, 'species', 'sepal_length', color='species', order=['virginica', 'versicolor', 'setosa'], 
            color_order=['virginica', 'versicolor', 'setosa'])
```

Activate fill. Change alpha.

```python
ax = violin(iris, 'species', 'sepal_length', color='species', fill=False, legend=None, alpha=1)
```

Color as the color variable.

```python
ax = violin(iris, 'species', 'sepal_length', color='species', fill=False, edgecolor=None)
```

Change orientation.

```python
ax = violin(iris, 'sepal_length', 'species', color='species', orient='h')
```

Add boxplots.

```python
ax = violin(iris, 'species', 'sepal_width', color='species', box=box_parameters(), alpha=0.5, legend=None)
```

Add boxplots and color as different value.

```python
ax = violin(iris, 'species', 'sepal_width', color='size', width=0.8, alpha=0.5, box=box_parameters())
```

Customize boxplot parameters.

```python
ax = violin(iris, 'species', 'sepal_width', color='species', legend=None, fill=False,
            box = box_parameters(fill_color='black', edge_color='red', edge_width=2, 
                                 median_color='red', median_width=3, outliers=True, 
                                 outlier_color='red', outlier_shape='x', outlier_size=10))
```

Change orientation.

```python
ax = violin(iris, 'sepal_length', 'species', color='species', legend=None, fill=False,
            box = box_parameters(fill_color='black', edge_color='red', edge_width=2, 
                                 median_color='red', median_width=3, outliers=True, 
                                 outlier_color='red', outlier_shape='x', outlier_size=10), orient='h')
```
