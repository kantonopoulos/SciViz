```python
import matplotlib.pyplot as plt
import seaborn as sns
from fun import histogram
```

```python
iris = sns.load_dataset('iris')
iris['size'] = iris['sepal_length'].apply(lambda x: 'big' if x > 5.5 else 'small')
```

Simple one variable histogram.

```python
ax = histogram(iris, 'sepal_length')
```

Increase number of bins.

```python
ax = histogram(iris, 'sepal_length', bins=30)
```

```python
ax = histogram(iris, 'sepal_length', binwidth=0.2)
```

Change color.

```python
ax = histogram(iris, 'sepal_length', bins=10, color_pal='green')
```

Add color variable.

```python
ax = histogram(iris, 'sepal_length', color='species', bins = 20, alpha=0.5)
```

Remove edge color and legend.

```python
ax = histogram(iris, 'sepal_length', color='species', bins = 20, edgecolor=None, legend=None)
```

Change color order.

```python
ax = histogram(iris, 'sepal_length', color='species', bins = 20, color_order=['virginica', 'versicolor', 'setosa'])
```

Change y axis statistic from 'count' to 'percentage'.

```python
ax = histogram(iris, 'sepal_length', stat='percent', bins=10)
```
