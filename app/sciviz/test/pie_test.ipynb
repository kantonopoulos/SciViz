```python
import matplotlib.pyplot as plt
import seaborn as sns
from fun import pie, label_parameters, text_parameters
```

```python
iris = sns.load_dataset('iris')
iris['size'] = iris['sepal_length'].apply(lambda x: 'big' if x > 5.5 else 'small')
```

Create simple piechart.

```python
ax = pie(iris, 'species')
```

Donut chart.

```python
ax = pie(iris, 'species', donut=True)
```

Add labels instead of legend. Add percentage in piechart.

```python
ax = pie(iris, 'size', labels=label_parameters(), text=text_parameters(), legend=None)
```

Change order.

```python
ax = pie(iris, 'size', order=['small', 'big'], labels=label_parameters(), text=text_parameters(), legend=None)
```

Change palette.

```python
ax = pie(iris, 'size', labels=label_parameters(), text=text_parameters(), legend=None, color_pal=['#FFFF80', '#603479'])
```

Change alpha and customize labels and text.

```python
ax = pie(iris, 'size', alpha=0.5, labels=label_parameters(size=14, color='blue'), 
         text=text_parameters(format='%1.2f%%', size=12, color='white'), legend=None)
```
