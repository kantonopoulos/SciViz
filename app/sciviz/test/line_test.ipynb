```python
iris = sns.load_dataset('iris')
iris['size'] = iris['sepal_length'].apply(lambda x: 'big' if x > 5.5 else 'small')
```

Simple lineplot with x and y variables.

```python
ax = sciviz.line(iris, 'sepal_width', 'sepal_length')
```

Add color variable. Remove legend.

```python
ax = sciviz.line(iris, 'sepal_width', 'sepal_length', color='species', legend=None)
```

Add shape variable same with color.

```python
ax = sciviz.line(iris, 'sepal_width', 'sepal_length', color='species', shape='species')
```

Add shape with another variable. Add color palette and size palette. Add also statistic to clean the plot.

```python
ax = sciviz.line(iris, 'sepal_width', 'sepal_length', color='size', shape='species', stat='mean',
          color_pal=['#FFFF80', '#603479'], shape_pal=['o', 's', 'X'])
```

Add group variable.

```python
ax = sciviz.line(iris, 'sepal_width', 'sepal_length', color='species', shape='species')
```

Change color and shape order. Change alpha.

```python
ax = sciviz.line(iris, 'sepal_width', 'sepal_length', color='species', shape='species', alpha=1,
          color_order=['virginica', 'versicolor', 'setosa'], shape_order=['virginica', 'versicolor', 'setosa'])
```

Add error bars.

```python
ax = sciviz.line(iris, 'sepal_width', 'sepal_length', color='size', stat='mean', errorbar=('ci', 95))
```

Change errorbar style.

```python
ax = sciviz.line(iris, 'sepal_width', 'sepal_length', color='species', stat='mean', errorbar=('ci', 95), errorbar_style='band')
```
