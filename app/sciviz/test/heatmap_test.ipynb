```python
import matplotlib.pyplot as plt
import seaborn as sns
from fun import heatmap, legend_parameters, tick_parameters
```

```python
iris = sns.load_dataset('iris')
iris['size'] = iris['sepal_length'].apply(lambda x: 'big' if x > 5.5 else 'small')
```

Simple heatmap just with data.

```python
ax = heatmap(iris, row_cluster=None, col_cluster=None)
```

No cbar.

```python
ax = heatmap(iris, row_cluster=False, col_cluster=False, cbar=False)
```

Simple heatmap just with data and clustering.

```python
ax = heatmap(iris)
```

Change gradient palette.

```python
ax = heatmap(iris, gradient_pal='coolwarm')
```

Uncluster cols.

```python
ax = heatmap(iris, gradient_pal='coolwarm', col_cluster=False)
```

Change dendrograms width.

```python
ax = heatmap(iris, gradient_pal='coolwarm', row_cluster=True, col_cluster=True, dendrogram=0.2)
```

Remove column dendrogram and clustering.

```python
ax = heatmap(iris, gradient_pal='coolwarm', row_cluster=True, col_cluster=False)
```

Remove dendrograms. Customize axis ticks.

```python
ax = heatmap(iris, gradient_pal='coolwarm', row_cluster=True, col_cluster=True, dendrogram=None, 
             ticks=tick_parameters(xticks=True, yticks=None, xticks_angle=90, ticklabel_size=15))
```

Add row annotation 1.

```python
ax = heatmap(iris, row_cluster=True, col_cluster=True, row1_annot='species')
```

Remove color bar.

```python
ax = heatmap(iris, row_cluster=True, col_cluster=True, row1_annot='species', cbar=False)
```

Add second row annotation. Customize legend.

```python
ax = heatmap(iris, row_cluster=True, col_cluster=True, row1_annot='species', row2_annot='size', row2_pal=['orange', 'pink'],
             legend=legend_parameters(orient='h', posx=0, posy=-0.6, title=['Species', 'Size'], title_size=13, title_bold=True))
```

Add col annotations.

```python
col_annot1 = [{'sepal_length': 'sepal', 'petal_width': 'petal', 'petal_length': 'petal', 'sepal_width': 'sepal'}, 'Part']
col_annot2 = [{'sepal_length': 'length', 'petal_width': 'width', 'petal_length': 'length', 'sepal_width': 'width'}, 'L/W']

ax = heatmap(iris, row_cluster=True, col_cluster=True, row1_annot='species', row2_annot='size', row2_pal=['orange', 'pink'],
            col1_annot=col_annot1, col1_pal=['black', 'grey'], col2_annot=col_annot2, col2_pal=['purple', 'green'],
            legend=legend_parameters(title=['Species', 'Size', 'Part', 'L/W'], title_size=13, title_bold=True))
```
