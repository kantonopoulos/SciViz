```python
import matplotlib.pyplot as plt
import seaborn as sns
from fun import boxplot, outlier_parameters, jitter_parameters
```

```python
iris = sns.load_dataset('iris')
iris['size'] = iris['sepal_length'].apply(lambda x: 'big' if x > 5.5 else 'small')
```

Simple boxplot with x and y variables.

```python
ax = boxplot(iris, 'species', 'sepal_length')
```

Add color variable.

```python
ax = boxplot(iris, 'species', 'sepal_length', color='species')
```

Add another color variable and color palette.

```python
ax = boxplot(iris, 'species', 'sepal_length', color='size', color_pal=['#FFFF80', '#603479'])
```

Change color order.

```python
ax = boxplot(iris, 'species', 'sepal_length', color='species', color_order=['virginica', 'versicolor', 'setosa'])
```

Change order and color order.

```python
ax = boxplot(iris, 'species', 'sepal_length', color='species', order=['virginica', 'versicolor', 'setosa'], 
         color_order=['virginica', 'versicolor', 'setosa'])
```

Change order and color order.

```python
ax = boxplot(iris, 'species', 'sepal_length', color='species', order=['virginica', 'versicolor', 'setosa'], 
         color_order=['virginica', 'versicolor', 'setosa'])
```

Change fill and width.

```python
ax = boxplot(iris, 'species', 'sepal_length', color='species', width=0.8, fill=False, edgecolor=None, legend=None)
```

Add black edgecolor and add caps.

```python
ax = boxplot(iris, 'species', 'sepal_length', color='species', fill=False, edgecolor='black', caps=True, legend=None)
```

Customize outliers.

```python
ax = boxplot(iris, 'species', 'sepal_length', color='species', outliers=outlier_parameters(color='red', shape='^', size=10))
```

Remove outliers.

```python
ax = boxplot(iris, 'species', 'sepal_length', color='species', outliers=None)
```

Add data points.

```python
ax = boxplot(iris, 'species', 'sepal_length', color='species', outliers=None, jitter=jitter_parameters(color_pal=None, size=50, alpha=0.8, pos='front'))
```

Make data points black.

```python
ax = boxplot(iris, 'species', 'sepal_length', color='species', outliers=None, jitter=jitter_parameters(color_pal=['black'], size=50, alpha=0.8, pos='front'))
```

Customize points and send them back.

```python
ax = boxplot(iris, 'species', 'sepal_length', color='species', outliers=None, alpha=0.9, jitter=jitter_parameters(jitter=0.05, color_pal=['black'], size=80, alpha=1, pos='back'))
```

Add points with color value different than x.

```python
ax = boxplot(iris, 'species', 'sepal_length', color='size', width=0.8, outliers=None, jitter=jitter_parameters(jitter=0.05, color_pal=['black']))
```

Change orientation.

```python
ax = boxplot(iris, 'sepal_length', 'species', color='size', width=0.8, outliers=None, orient='h', jitter=jitter_parameters(jitter=0.05, color_pal=['black']))
```
